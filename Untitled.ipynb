```python
import pandas as pd
from pathlib import Path
import numpy as np
%matplotlib inline
```

```python
# 2 import and read data set 
```

```python
drought_data = Path("us_state_data.csv")

drought_df = pd.read_csv(drought_data)

drought_df.head(10)
```

```python
drought_df.dtypes
```

```python
drought_df.set_index(pd.to_datetime(drought_df["MapDate"], infer_datetime_format = True),inplace = True)
drought_df.tail(15)
```

```python
drought_df.drop("MapDate", axis = 1, inplace = True)
drought_df.head()
```

```python
drought_df = drought_df.drop(columns = ["PercentChangeFromWAve", "AreaChangeFromWAve", "StatisticFormatID"])
drought_df.tail(20)
```

```python
all_states_df = drought_df.groupby( ["Name","AreaCurrentPercent", "PopulationCurrentPercent","USDMLevel"])['Name'].count()
all_states_df.tail(50)
```

```python

drought_percent = drought_df.iloc[:, [1,5,6,7]]
drought_percent
```

```python
droughtlevel_per_state = drought_percent.groupby('Name').agg(av_drought_level =('USDMLevelID', 'mean'), av_drought_area = ('AreaCurrentPercent' , 'mean'))
sorthed_drought_df = droughtlevel_per_state.sort_values("av_drought_level", ascending=False)
```

```python
sorthed_drought_df
```
